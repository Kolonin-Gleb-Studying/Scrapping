```python
inputAdress = "24.48.0.1"
# расскоментировать для ручного ввода
inputAdress = input("Введите IP-адрес: ") # 24.48.0.1
```

```python
!pip install pycurl
import pycurl
from io import BytesIO # Библиотека для работы с байтовыми последовательностями
```

```python
# Создаем объект для приема данных в формате последовательности байтов
b_obj = BytesIO()

# Создаем объект Curl
crl = pycurl.Curl()

# Параметры для запроса данных
crl.setopt(crl.URL, f'http://ip-api.com/json/{inputAdress}')

# Следуем за редиректами
crl.setopt(crl.FOLLOWLOCATION, True)

# Устанавливаем объект, в который будет направлен поток данных, который вернет сервер
crl.setopt(crl.WRITEDATA, b_obj)

# Выполняем запрос
crl.perform()

# Закрываем соединение
crl.close()

# Получаем содержимое в виде строки
get_body = b_obj.getvalue()

print(get_body)
```

```python
import json
import pandas as pd

ip_dict = json.loads(get_body)
keys = ["Статус", "Страна", "Код_Страны", "Код_Региона", "Регион", "Город", "Почтовый_индекс", "Широта", "Долгота", "Временная_зона", "isp", "org", "as", "ip_адрес"]

ip_dict = dict(zip(keys, ip_dict.values()))
ip_dict
```

```python
# Подключаем библиотеки для работы с выводом в Colab
from IPython.display import display, HTML, IFrame
```

```python
# Формирование строки заголовков

headers = "<tr>"
for col in ip_dict.keys():
    headers += "<th>"+col+"</th>\n"
headers += "</tr>"

headers
```

```python
# Формирование строк таблицы.
# Выведем по очереди все строки и столбцы в них
table = ''

for val in ip_dict.values():
    table += "<td>"+str(val)+"</td>"

# Отображение заголовков + строк в виде единой HTML таблицы
display(HTML("<h1>Информация о IP</h1><table border=1 cellpadding=10>"+headers+table+"</table>"))
```
