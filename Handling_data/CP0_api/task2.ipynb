```python
# 3) Используя API https://quickchart.io/documentation/ постройте два графика (можете каждый на своём полотне, можете все два на одном):
# - Аренда торговых площадей
# - Аренда офисов
# Все графики сохраните в файлы и отобразите в колабе
```

```python
# - Аренда торговых площадей
# Как менялось число арендованных торговых площадей
# Как менялось число офисных площадей
```

```python
# Выжимка нужных данных для графиков
dates = df_html[1]["Месяц"].tolist()
trade_areas = df_html[1]["Аренда торговых площадей₽/м²/год"].tolist()
trade_areas = [float(val.replace(" ", "")) for val in trade_areas]
ofice_areas = df_html[1]["Аренда офисов₽/м²/год"].tolist()
ofice_areas = [float(val.replace(" ", "")) for val in ofice_areas]
```

```python
print(dates)
print(trade_areas)
print(ofice_areas)
```

```python
# Нужные библиотеки для использования API
!pip install pycurl

import pycurl
from io import BytesIO # Библиотека для работы с байтовыми последовательностями
```

```python
import json

# В этот файл запишем ответ
f = open("trade_areas.png", "wb")

# Подготовим POST-парамеры
data_chart = {
  'chart': {
      'type': 'line', 
      'data': {
          'labels': dates, # Это годы
          'datasets': [{
              'label': 'Аренда торговых площадей₽/м²/год',                           
              'data': trade_areas  # данные о площади
          }]
      }
    }
}

data = json.dumps(data_chart)
print(data)

#data_post = urlencode(data)

b_obj = BytesIO()
crl = pycurl.Curl()
crl.setopt(crl.URL, 'https://quickchart.io/chart')
crl.setopt(crl.FOLLOWLOCATION, True)
crl.setopt(crl.WRITEDATA, f)
crl.setopt(crl.HTTPHEADER, ["Content-Type: application/json"])
crl.setopt(crl.POSTFIELDS, data)

crl.perform()
crl.close()

f.close()

print(b_obj)
```
